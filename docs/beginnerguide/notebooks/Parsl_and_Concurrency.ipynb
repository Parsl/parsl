```python
# %% [markdown]
# ## Import Parsl and Other Necessary Libraries

# %%
# Install Parsl
!pip install parsl

import parsl
from parsl import python_app, HighThroughputExecutor
from parsl.config import Config

# %% [markdown]
# ## Configure Parsl

# %%
# Define a configuration that specifies the use of the HighThroughputExecutor.
# This executor is suitable for scaling across many compute nodes.

config = Config(
    executors=[HighThroughputExecutor(
        label="htex_Local",
        worker_debug=True,
        cores_per_worker=1,
        max_workers=4  # Adjust based on your local machine's resources
    )]
)

parsl.load(config)

# %% [markdown]
# ## Define Parsl Python Apps

# %%
# This decorator transforms the function into a Parsl app that can run concurrently.
@python_app
def multiply(x, y):
    return x * y

@python_app
def add(x, y):
    return x + y

# %% [markdown]
# ## Execute Apps Asynchronously

# %%
# Run the apps with various inputs
result1 = multiply(5, 10)
result2 = add(20, 30)

# %% [markdown]
# ## Retrieve and Print Results

# %%
# Use the `result()` method to retrieve the output once computation is complete.
print("Result of multiplication:", result1.result())
print("Result of addition:", result2.result())
```
